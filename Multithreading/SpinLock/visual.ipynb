```python
import pandas as pd
import numpy as np
ttas = pd.read_csv("ttas_data.csv", header=None)
ticket = pd.read_csv("ticket_data.csv", header=None)
```

```python
ttas
```

```python
x = np.arange(1,163,2)    
ttas_max =       ttas.iloc[0::3, 0]
ttas_avg =       ttas.iloc[0::3, 1]
ttas_exp_max =   ttas.iloc[1::3, 0]
ttas_exp_avg =   ttas.iloc[1::3, 1]
ttas_yield_max = ttas.iloc[2::3, 0]
ttas_yield_avg = ttas.iloc[2::3, 1]
 
ticket_max =       ticket.iloc[0::3, 0]
ticket_avg =       ticket.iloc[0::3, 1]
ticket_exp_max =   ticket.iloc[1::3, 0]
ticket_exp_avg =   ticket.iloc[1::3, 1]
ticket_yield_max = ticket.iloc[2::3, 0]
ticket_yield_avg = ticket.iloc[2::3, 1]
```

```python
import matplotlib.pyplot as plt

plt.plot(x,ttas_max, color='r', label = 'ttas max')
plt.plot(x,ttas_avg, color='r', label = 'ttas avg')
plt.plot(x,ttas_exp_max, color='g', label = 'ttas_exp_back_off max')
plt.plot(x,ttas_exp_avg, color='g', label = 'ttas_exp_back_off avg')
plt.plot(x,ttas_yield_max, color='b', label = 'ttas_yield max')
plt.plot(x,ttas_yield_avg, color='b', label = 'ttas_yield avg')
plt.legend()
plt.xlabel('thread number')
plt.ylabel('time to lock, milliseconds')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = np.arange(1,101,2)
plt.plot(x,ticket_max, color='r', label = 'ticket max')
plt.plot(x,ticket_avg, color='r', label = 'ticket avg')
plt.plot(x,ticket_exp_max, color='g', label = 'ticket_exp_back_off max')
plt.plot(x,ticket_exp_avg, color='g', label = 'ticket_exp_back_off avg')
plt.plot(x,ticket_yield_max, color='b', label = 'ticket_yield max')
plt.plot(x,ticket_yield_avg, color='b', label = 'ticket_yield avg')
plt.legend()
plt.xlabel('thread number')
plt.ylabel('time to lock, milliseconds')
plt.show()
```
