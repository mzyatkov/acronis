```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("onethread_naive.dat")
data[:].plot.scatter(0,1,10).get_figure().savefig('onethread')
```

```python
data = pd.read_csv("onethread_naive.dat")
data2 = pd.read_csv("onethread_block.dat")
X1 = data["matrix_size"]
Y1 = data["time_sec"]
X2 = data2["matrix_size"]
Y2 = data2["time_sec"]

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(X1, Y1, s=10, c='b', marker="s", label='naive')
ax1.scatter(X2,Y2, s=10, c='r', marker="o", label='block (8x8)')
plt.legend()
plt.show()
```

```python
data3 = pd.read_csv("multithread_naive.dat")
data4 = pd.read_csv("multithread_block.dat")

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_title("graph for 2048*2048")
ax1.scatter(data3.iloc[:,0], data3.iloc[:,1], s=10, c='b', marker="s", label='naive')
ax1.scatter(data4.iloc[:,0], data4.iloc[:,1], s=10, c='r', marker="s", label='block (16x16)')
ax1.set_ylabel("times, sec")
ax1.set_xlabel("threads number")
ax1.set_yscale('log')
plt.legend()
plt.show()
plt.savefig("multithread2048,batch16")
```

```python
thr_num = 8
print(f'ускорение для наивного в {thr_num} потоков = {data3.iloc[0,1]/data3.iloc[thr_num-1,1]} ')
print(f'ускорение для блочного в {thr_num} потоков = {data4.iloc[0,1]/data4.iloc[thr_num-1,1]} ')
```
