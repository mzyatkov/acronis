```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

simple_locked = pd.read_csv('simple_locked.csv', header=None)
random_locked = pd.read_csv('random_locked.csv', header=None)
unbalanced_locked = pd.read_csv('unbalanced_locked.csv', header=None)
simple_lockfree = pd.read_csv('simple_lockfree.csv', header=None)
random_lockfree = pd.read_csv('random_lockfree.csv', header=None)
unbalanced_lockfree = pd.read_csv('unbalanced_lockfree.csv', header=None)

def plot_graph(df, title, xlabel, ylabel, color):
    plt.plot(df[0], df[1], color, label=title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

plt.title('Locked Stack')
plot_graph(simple_locked, 'Simple Lock', 'Number of Threads', 'Time (ms)', 'r')
plot_graph(simple_locked, 'Random Lock', 'Number of Threads', 'Time (ms)', 'g')
plot_graph(simple_locked, 'Unbalanced Lock', 'Number of Threads', 'Time (ms)', 'b')
plt.show()

plt.title('Lock Free Stack')
plot_graph(simple_lockfree, 'Simple LockFree', 'Number of Threads', 'Time (ms)', 'r')
plot_graph(random_lockfree, 'Random LockFree', 'Number of Threads', 'Time (ms)', 'g')
plot_graph(unbalanced_lockfree, 'Unbalanced LockFree', 'Number of Threads', 'Time (ms)', 'b')
plt.legend()
plt.show()
```
