```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(pd.read_csv("ms_bucket.csv")).dropna()
timestamp = data[data.columns[0]]
inf = data[data.columns[1]]
data = data[sorted(data.columns[2:].tolist(), key = lambda x: int(x))]
```

```python
data
```

```python
for i in range(data.shape[0]):
    data.iloc[i]=data.iloc[i]/inf[i] #нормируем функцию распределения
```

```python
data
```

$${\displaystyle \mathbb {E} [X]=\int \limits _{-\infty }^{\infty }\!x\,dF_{X}(x)}, {\displaystyle x\in \mathbb {R} }\\{\displaystyle D[X]=E\left[{\big (}X-E[X]{\big )}^{2}\right]}$$


```python
x_times = list(map(int, data.columns.tolist()))
x_times = np.array([0] + x_times) #координаты по оси x
rownum = data.shape[0]
exp_value = [0]*rownum
dispersion = [0]*rownum
for i in range(rownum):
    row = np.array([0] + data.iloc[i].tolist())
    f = np.gradient(row,x_times) # плотность распределения
    exp_value[i] = np.trapz(f*x_times,x_times) #матожидание
    dispersion[i] = np.trapz(f*np.square(x_times - exp_value[i]),x_times) #дисперсия
```

```python
plt.plot(timestamp, exp_value)
plt.show()
plt.plot(timestamp, dispersion)
plt.show()
```
